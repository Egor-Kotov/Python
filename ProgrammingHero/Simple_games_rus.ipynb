```python
"""Угадай число от 1 до 10"""


from random import randint


points = 0


while True:
    rand = randint(1, 10)
    number = input('Введите число от 1 до 10\nИли q для выхода из игры ')
    if number == 'q': 
        points = 0
        break
    elif rand == int(number):
        points +=10
        print(f"\nКрасавчик!!!\nТвой счет: {points}\nЕщё разок???\n")
    else:
        print(f'\nО, нет!! Это было {rand}\nЕщё разок?')
```

```python
"""Камень-ножницы-бумага"""


from random import randint


def igra():
    evil = 0
    good = 0
    
    
    def chiu(x, y):
        if x > y:
            return "в твою пользу, добивай уже!))"
        elif y > x:
            return "редиска побеждает((("
        else:
            return "ничья... поднажми!"
    
    
    print("\nПогнали брат, до трех побед!\n")
    
    
    while True:
        if  evil == 3:
            print("\n\n\nУнижен и растоптан... Зачем ты вообще начал, если ни на что неспособен...")
            good =  evil = 0
            break
        elif good == 3:
            print("\n\n\nКрасавчик! Если бы знал, что ты так хорош, выбрал тебе соперника помощнее. Следующего сюда несите!")
            good = evil = 0
            break
        
        vibor = input('\nКамень, ножницы, бумага, цу-е-фа! ')
        vibor = vibor.lower()
        
        if vibor != 'камень' and vibor != 'бумага' and vibor != 'ножницы':
            print("Опечатка, напиши нормально!")
            continue 
        
        protiv = randint(1, 3)
        
        if protiv == 1:
            protiv = 'камень'
        elif protiv == 2:
            protiv = 'ножницы'
        else:
            protiv = 'бумага'
  
        if vibor == 'камень' and protiv == 'ножницы' or vibor == 'ножницы' and protiv == 'бумага' or vibor == 'бумага' and protiv == 'камень':
            good += 1
            print(f"Ты разматал, у противника - {protiv}.")
            print(f"Счет {good} - {evil} {chiu(good, evil)}")
        elif vibor == protiv:
            print(f"Пффф... {vibor} - {protiv}\nЕщё разок!")
        else:
            evil += 1
            print(f"Это ловушка!!! У противника - {protiv}.")
            print(f"Счет {good} - {evil} {chiu(good, evil)}")

raz = 0


def razok():
    if raz == 0:
        return ''
    else:
        return 'еще разок '             

    
while True:
    a = input(f'\nХочешь {razok()}сыграть в камень-ножницы-бумага???\nВведи да или нет: ')
    a = a.lower()
    
    if a != 'да' and a != 'нет':
        print('\nПиши пожалуйста, по-русски...')
        continue
    
    if a == 'да':
        raz += 1
        igra()
    else:
        raz = 0
        print('Трусишь небось... Ладно мб в другой раз')
        break
```

```python
"""Быки и коровы"""

from random import randint


while True:
    try:
        n = int(input("\nСколько цифр в числе?\n(2, 3 или 4) "))
    except:
        print("\nВведите цифру")
        continue
    if n == 3 or n == 2 or n == 4:
        break
    else:
        print("\nВыберите из доступных вариантов")

        
def unuqint_randlist(x):
    a = []
    a.append(randint(1, 9))
    while len(a) < x:
        i = randint(0, 9)
        if i not in a:
            a.append(i)
    return a


n = unuqint_randlist(n)


def slova(x):
    if x == 2:
        return "от 10 до 99"
    elif x == 3:
        return "от 100 до 999"
    else:
        return "от 1000 до 9999"

def u_l(x):
    y = x[:]
    z = y.pop()
    if z in y:
        return True

while True:
    try:
        popitka = int(input("\nЗа сколько попыток управишься? "))
        break
    except ValueError:
        print("Вы ввели число? Попытайтесь еще раз.")
        continue
    

while True:
    if popitka == 0:
        print(f"\nУ вас закончились попытки((((\nЭто было число {''.join([str(i) for i in n])}\nGame Over")
        break
    try:
        dogadka = int(input("\nКакое число, как думаешь? "))
    except ValueError:
        print("Вы ввели число? Попытайтесь еще раз.")
        continue
    dogadka = [int(i) for i in str(dogadka)]
    
    if len(n) != len(dogadka):
        print(f"Мы загадали число {slova(len(n))}")
        continue
    elif u_l(dogadka):
        print("Число должно быть с неповторяющимися цифрами")
        continue
    elif dogadka == n:
        print('Вы угадали, умница!')
        break
    
    cow = 0
    bull = 0
    
    for i in range(len(n)):
        if dogadka[i] == n[i]:
            bull += 1
        elif dogadka[i] in n:
            cow += 1
    
    popitka -= 1
    print(f"{bull} быков и {cow} коров, осталось {popitka} попыток")
```

```python
"""Угадай слово"""


import requests, random
from bs4 import BeautifulSoup


"""Создаю список слов (1000 самых употребляемых существительных)"""
url = 'http://dict.ruslang.ru/freq.php?act=show&dic=freq_s&title=%D7%E0%F1%F2%EE%F2%ED%FB%E9%20%F1%EF%E8%F1%EE%EA%20%E8%EC%E5%ED%20%F1%F3%F9%E5%F1%F2%E2%E8%F2%E5%EB%FC%ED%FB%F5'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

slova = []

for string in soup.stripped_strings:
    slova.append(string)
slova = (slova[12::4])[:-2]
points = 0


"""Создаю игру"""
def igra_slov():
    global points
    start = input("\nХочешь сыграть в 'Заверши слово'?\n(Выбери да или нет) ")
    if start == 'да':
        print(f"\nПогнали!\nУ тебя {points} очков\nЗагадано существительное.")
    elif start == "нет":
        print(f'\nКак знаешь, ты заработал {points} очков!\nПока-пока!')
        points = 0
        return None
    else:
        return igra_slov()
# выбираем слово    
    r = random.randint(0, len(slova)-1)
    zagadka = list(slova[r])
# превращаем в ребус, l = количество скрытых букв
    l = abs(-len(zagadka)//2)
    
    def rebus(x, y):
        if x.count("-") == y:
            return x
        x[random.randint(0, len(x)-1)] = "-"
        return rebus(x, y)
    
    zagadka = rebus(zagadka, l)
    zagadka = "".join([i for i in zagadka])

# Запускаем петлю, которая просит ответ
    while True:
        otgatka = input(f"\nУгадайка ? {zagadka} ? ")
        
        if otgatka in slova:
            print("\n\nУмница, это победа!")
            points += 10
            break
        else:
            print("\nМимо((( Еще раз!")
            points -= 1
            continue
    return igra_slov()   


igra_slov()
```
